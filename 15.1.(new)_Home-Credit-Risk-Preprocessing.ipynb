# **CHECKING**

***Xóa đi các columns chứa quá nhiều giá trị null (>60%)***

```python
remove_null_cols = []
for i in data.columns:
  if data[i].isnull().sum() > int(data.shape[0]*0.6):
    remove_null_cols.append(i)
print(remove_null_cols)
print(len(remove_null_cols))
```

```python
data = data.drop(columns = remove_null_cols)
data.shape
```

***Một số columns sẽ được loại đi tiếp theo, vì những cột này không cần thiết và đã có những cột khác có vai trò tương tự:***

```python
data = data.drop(columns = ['EXT_SOURCE_1', 'EXT_SOURCE_3', 'APARTMENTS_MEDI', 'BASEMENTAREA_MEDI', 'YEARS_BEGINEXPLUATATION_MEDI',
                            'ELEVATORS_MEDI', 'ENTRANCES_MEDI', 'FLOORSMAX_MEDI', 'LANDAREA_MEDI', 'LIVINGAREA_MEDI', 'NONLIVINGAREA_MEDI'])
```

***Có một số cột chứa nhiều giá trị Null, ta sẽ bù đắp các giá trị Null đó bằng cách: "Nhóm" các giá trị của cột này dựa trên một cột có liên quan, sau đó chọn ra giá trị nhiều nhất trong nhóm đó***

```python
#Cột OCCUPATION_TYPE sẽ được bù Null dựa vào NAME_INCOME_TYPE:
mode_by_OCCUPATION_TYPE = data.groupby("NAME_INCOME_TYPE")['OCCUPATION_TYPE'].agg(lambda x: x.mode()[0] if not x.mode().empty else np.nan)
data['OCCUPATION_TYPE'] = data['OCCUPATION_TYPE'].fillna(data["NAME_INCOME_TYPE"].map(mode_by_OCCUPATION_TYPE))
```

```python
#Tương tự, các cột _AVG sẽ được tính trung bình dựa vào NAME_HOUSING_TYPE (loại nhà)
AVG_ls = ['APARTMENTS_AVG', 'BASEMENTAREA_AVG', 'YEARS_BEGINEXPLUATATION_AVG', 'ELEVATORS_AVG', 'ENTRANCES_AVG',
          'FLOORSMAX_AVG', 'LANDAREA_AVG', 'LIVINGAREA_AVG', 'NONLIVINGAREA_AVG']
for i in AVG_ls:
  data.loc[:, i] = data[i].fillna(data.groupby('NAME_HOUSING_TYPE')[i].transform('mean'))
```

```python
# Tương tự với các cột _MODE, tính như _AVG
MODE_ls = ['APARTMENTS_MODE', 'BASEMENTAREA_MODE', 'YEARS_BEGINEXPLUATATION_MODE', 'ELEVATORS_MODE', 'ENTRANCES_MODE',
           'FLOORSMAX_MODE', 'LANDAREA_MODE', 'LIVINGAREA_MODE', 'NONLIVINGAREA_MODE', 'TOTALAREA_MODE']
for i in MODE_ls:
  data.loc[:, i] = data[i].fillna(data.groupby('NAME_HOUSING_TYPE')[i].transform('mean'))
```

```python
#Vì 3 cột này mang định dạng Object, nên ta sử dụng phương pháp chọn ra giá trị nhiều nhất
for i in ['HOUSETYPE_MODE', 'WALLSMATERIAL_MODE', 'EMERGENCYSTATE_MODE']:
  mode_values = data.groupby('NAME_HOUSING_TYPE')[i].transform(lambda x: x.mode()[0] if not x.mode().empty else pd.NA)
  data[i] = data[i].fillna(mode_values)
```

# **FEATURES SELECTION**

```python
for i in data.columns:
  print(i, len(data[i].unique()))
```

```python
numeric_cols = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE', 'REGION_POPULATION_RELATIVE', 'DAYS_BIRTH',
                'DAYS_EMPLOYED', 'DAYS_REGISTRATION', 'DAYS_ID_PUBLISH', 'EXT_SOURCE_2', 'APARTMENTS_AVG', 'BASEMENTAREA_AVG',
                'YEARS_BEGINEXPLUATATION_AVG', 'ELEVATORS_AVG', 'ENTRANCES_AVG', 'FLOORSMAX_AVG', 'LANDAREA_AVG', 'LIVINGAREA_AVG',
                'NONLIVINGAREA_AVG', 'APARTMENTS_MODE', 'BASEMENTAREA_MODE', 'YEARS_BEGINEXPLUATATION_MODE', 'LANDAREA_MODE',
                'LIVINGAREA_MODE', 'NONLIVINGAREA_MODE', 'TOTALAREA_MODE', 'DAYS_LAST_PHONE_CHANGE']
categoy_cols = data.columns.difference(numeric_cols)
```

***Chi-squared cho biến phân loại***

```python
remove_obj_cols =[]

#Chi-square test cho các cột phân loại so với cột Output của data
for i in categoy_cols :
  if i == 'TARGET':
    continue
  else:
    chi2, pval, _, _ = chi2_contingency(pd.crosstab(data[i], data['TARGET']))
    print(i, '---', pval)
    if pval > 0.05:
      remove_obj_cols.append(i)
    else:
      continue
print(remove_obj_cols)
```

```python
data = data.drop(columns = remove_obj_cols)
data.shape
```

***Spearman Correlation cho biến liên tục***

```python
spearman_df = data[numeric_cols + ['TARGET']]
correlation_results = []

# Duyệt qua tất cả các cột trong DataFrame
for col in spearman_df.columns:
  if col == 'TARGET':
    continue
  else:
    if spearman_df[col].nunique() > 1:
      corr, p_value = spearmanr(spearman_df[col], spearman_df['TARGET'])
      correlation_results.append((col, corr, p_value))
    else:
        print(f"Cột {col} có giá trị không thay đổi và bị bỏ qua.")

# Chuyển kết quả thành DataFrame để dễ quan sát
correlation_df = pd.DataFrame(correlation_results, columns=['Feature', 'Spearman Correlation', 'p-value'])
correlation_df = correlation_df.sort_values(by='Spearman Correlation', ascending=False)
correlation_df
```
